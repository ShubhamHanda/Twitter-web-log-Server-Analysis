```python
def Q1_created_deleted(jsonPathfile):
    numCreated = 0
    numDeleted = 0
    numTotal = 0
    numErrors = 0
    with open(jsonPathfile, "r") as jsonFile:
        for record in jsonFile:
            tweet = json.loads(record)
            if "created_at" in tweet:
                numTotal += 1
                if "delete" in tweet:
                    numErrors += 1
                else:
                    numCreated += 1
            elif "delete" in tweet:
                numTotal += 1
                if "created_at" in tweet:
                    numErrors += 1
                else:
                    numDeleted += 1
    return(numCreated,numDeleted,numTotal,numErrors)
```

```python
total_Created=0
total_Deleted=0
Total_Count=0
total_Error=0
for hour in hours:
    destination = os.path.join(dataDirectory, hour)
    print(hour, "\t", destination)
    jsonPathfile = r"C:\Users\Dell\Desktop\web2\2018\10\10" + '\\' + hour + '\\'
    files = os.listdir(destination)
    for file in files:
        if file.endswith(".json"):
            jsonPathfile = jsonPathfile + file
            Num_Created,Num_Deleted,Num_Total,Num_Errors = Q1_created_deleted(jsonPathfile)
            total_Created += Num_Created
            total_Deleted += Num_Deleted
            Total_Count += Num_Total
            total_Error += Num_Errors
            jsonPathfile = r"C:\Users\Dell\Desktop\web2\2018\10\10" + '\\' + hour + '\\'
            
```

```python
#q4 hashtags
def q4_hastags(jsonPathfile):  
    HashtagsFrequencyDist = defaultdict(int)
    with open(jsonPathfile, "r") as jsonFile:
        for record in jsonFile:
            tweet = json.loads(record)
            entities = tweet.get("entities", {})
            hashtags = entities.get("hashtags", {})
            counter_hashtags = len(hashtags)
            HashtagsFrequencyDist[counter_hashtags] += 1
    return HashtagsFrequencyDist
```

```python
#calculating usernames
def q5_username(jsonPathfile):    
    username_num = Counter()
    usermentions_list=[]
    with open(jsonPathfile, "r") as jsonFile:
        for record in jsonFile:
            tweet = json.loads(record)
            entities = tweet.get("entities", {})
            user_mentions = entities.get("user_mentions", {})
            usermentions_list = [tag["screen_name"] for tag in user_mentions]
            username_num.update(usermentions_list)
            
    return(username_num)
```

```python
def tokenization_words(jsonPathfile):  
    stop_words = set(stopwords.words('english')) 
    stop_words_list=list(stop_words)
    stop_words_list.append("rt")
    stop_words_list.append("via")
    stop_words_list.append(bin(10000101))
    text_number = Counter()
    Wordlist=[]
    tweettokenizer= TweetTokenizer()
    hashtagTextFrequencyDistribution = Counter()
    with open(jsonPathfile, "r") as jsonFile:
        for record in jsonFile:
            tweet = json.loads(record)
            if 'text' in tweet:
                text = tweet.get("text")
                text = tweettokenizer.tokenize(text.lower())
                text = [w for w in text if not w in stop_words_list]
                text = [w for w in text if w.isalpha()]
                text_number.update(text)
        return text_number
```
